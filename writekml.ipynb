```python
import pandas as pd
irr = pd.read_csv('data/irr.csv')
al = pd.read_csv('data/al.csv')
jm = pd.read_csv('data/jm.csv')

with open('sample.kml', 'w+') as the_file:
    the_file.write('<?xml version="1.0" encoding="UTF-8"?>\n')
    the_file.write('<kml xmlns="http://earth.google.com/kml/2.1">\n')
    the_file.write('  <Document>\n')
    the_file.write('    <name>NYC Traffic Jams</name>\n')
    the_file.write('    <description>NYC Traffic Jams</description>\n')
    the_file.write('\n')
    the_file.write('    <Style id="redLine">\n')
    the_file.write('      <LineStyle>\n')
    the_file.write('        <color>ff0000ff</color>\n')
    the_file.write('        <width>4</width>\n')
    the_file.write('      </LineStyle>\n')
    the_file.write('    </Style>\n')
    
    for index, row in jm.iterrows(): 
        if row['city'] == "Staten Island":
            the_file.write('          <Placemark>\n')
            the_file.write('          <name>Red Line</name>\n')
            the_file.write('          <styleUrl>#redLine</styleUrl>\n')
            the_file.write('          <LineString>\n')
            the_file.write('            <altitudeMode>relative</altitudeMode>\n')
            the_file.write('            <coordinates>\n')
    
            tokens=row['line'].split("}")
            for word in tokens: 
                the_file.write(word.replace("'y':","").replace("'x':","").replace("[{ ","").replace(", { ","").replace("]",""))
                the_file.write('\n')
     
            the_file.write('            </coordinates>\n')
            the_file.write('          </LineString>\n')
            the_file.write('          </Placemark>\n')
    
    the_file.write('      </Document>\n')
    the_file.write('    </kml>\n')
```
