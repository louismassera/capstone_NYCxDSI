```python
complaints_to_look_for = ["Illegal Parking", "Blocked Driveway", "Street Sign - Damaged", "Street Sign - Missing", "Street Condition", "Street Sign - Dangling", "Sewer", "Street Light Condition", "Traffic Signal Condition", "Broken Parking Meter", "Root/Sewer/Sidewalk Condition", "Highway Condition"]
```

```python
relevant_calls = calls.loc[calls['Complaint Type'].isin(complaints_to_look_for)]
```
