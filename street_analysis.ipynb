```python
import pandas as pd
```

```python
al = pd.read_csv("data/al_complete.csv")
```

```python
import datetime

al['pubMillis'] = al['pubMillis']/1000
al['date'] = al['pubMillis'].apply(lambda s: datetime.datetime.fromtimestamp(s).strftime('%Y-%m-%d %H:%M:%S'))
al['date'] = pd.to_datetime(al['date'])
```

```python
hazard_shoulder = al.loc[al['subtype'] == 'HAZARD_ON_SHOULDER_CAR_STOPPED']
```

```python
hazard_shoulder
```

```python
hazard_group_by_street = hazard_shoulder['street'].groupby(hazard_shoulder['street']).count()
```

```python
print (len(hazard_group_by_street))
hazard_group_by_street = hazard_group_by_street.sort_values()
print (hazard_group_by_street)
```

```python
jam_heavy = al.loc[al['subtype'] == 'JAM_HEAVY_TRAFFIC']
```

```python
jam_heavy_group_by_street = jam_heavy['street'].groupby(jam_heavy['street']).count()
```

```python
print (len(jam_heavy_group_by_street))
jam_heavy_group_by_street = jam_heavy_group_by_street.sort_values()
print (jam_heavy_group_by_street)
```

```python
hazard_group_by_street_top = hazard_group_by_street.tail(50)
jam_heavy_group_by_street_top = jam_heavy_group_by_street.tail(50)
```

```python
common_top_streets = []
for i in range(len(hazard_group_by_street_top.index)):
    if (hazard_group_by_street_top.index[i] in jam_heavy_group_by_street_top.index):
        common_top_streets.append(hazard_group_by_street_top.index[i])
```

```python
print (len(common_top_streets))
for i in range(len(common_top_streets)):
    print (common_top_streets[i])
```
